```python
#can we create new features from display_inches, resolution_height and resolution_width??
#total_pixels
#ppi
#aspect_ratio
df['ppi']=np.sqrt(df['resolution_width']**2+df['resolution_height']**2)/(df['resolution_width']*df['resolution_height'])
```

There is a negative correlation between (ppi and price) and (ppi and rating), indicating that as the ppi increases, both the price and rating tend to decrease

```python
#Are there any specific combinations of GPU and CPU that are associated with higher-priced or higher-rated laptops?
cpu_gpu=df.groupby(['cpu_brand','gpu_brand']).agg({'price':'mean','rating':'mean'}).round(2)
cpu_gpu
```

```python
cpu_gpu=cpu_gpu.sort_values(by='price',ascending=False)
cpu_gpu
```

```python
cpu_gpu['price'].plot(kind='bar',figsize=(12,6))
plt.xlabel('GPU-CPU Combinations')
plt.ylabel('Average Price')
plt.title('Top GPU-CPU Combinations Associated with Higher-Priced Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Yes! NVIDIA and Intel are gpu-cpu combination which is costliest

```python
cpu_gpu['rating'].plot(kind='bar')
plt.xlabel('GPU-CPU Combinations')
plt.ylabel('Average Rating')
plt.title('Top GPU-CPU Combinations Associated with Higher-Rated Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ram_rom=df.groupby(['ram_gb','rom_gb']).agg({'price':'mean','rating':'mean'})
ram_rom=ram_rom.sort_values(by='price',ascending=False)
ram_rom
```

```python
ram_rom['price'].plot(kind='bar',figsize=(12,6))
plt.xlabel('ram_rom combination')
plt.ylabel('Avg_price')
plt.title('Distribution of ram_rom combination associate with Avg_price')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
ram_rom['rating'].plot(kind='bar',figsize=(12,6))
plt.xlabel('ram_rom combination')
plt.ylabel('Avg_price')
plt.title('Distribution of ram_rom combination associate with rating')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
